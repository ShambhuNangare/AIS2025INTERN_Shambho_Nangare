# Box and Whisker Plot (Boxplot)

* A boxplot visually represents data dostribution.
* The box represent Q1 (25th percentile), Q3 (75th percentile), and the median (50th percentile).
* The wisker extend to 1.5 * IQR.
* Any dots outside the whiskers are outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("Titanic_dataset.csv")
```

```python
plt.figure(figsize = (6,4))
sns.boxplot(y = df["Fare"])
plt.title("Boxplot of Titanic Fare")
plt.show()
```
