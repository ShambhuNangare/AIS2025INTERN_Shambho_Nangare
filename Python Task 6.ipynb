# Matplotlib Basics

Matplotlib is a comprehensive plotting library for Python. It is widely used for creating static, animated, and interactive visualizations in Python. Matplotlib can be used to create a wide range of plots, such as line plots, scatter plots, bar charts, histograms, 3D plots, etc. The library is highly customizable, allowing you to change every aspect of a plot.

Matplotlib is often used in data analysis, data visualization, scientific computing, and machine learning.

* Key Concepts and Components  
Figure: The whole window or image containing one or more axes (subplots). It's the top-level container for all elements.  
Axes: A single plot or graph inside the figure. It contains the plot, the labels, the title, the grid, and the axis ticks.  
Axis: The x and y axes (or 3D axes in some cases) of a plot.

* LINE PLOT  
  This is often used to visualize trends over time or a relationship between two continuous variables.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.title('Line plot example')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.show()
```

* Interpretation   

  This plot shows the relationship between x and y, where y increases as x increases.  
  This plot represents a quadratic relationship since the value of y follow the pattern of y = x^2.  
  The markers (0) help highlight individual data points on the line.
  plt.plot(x, y) creates a line plot with x-values and y-values.
  plt.title(), plt.xlabel(), and plt.ylabel() are used to set the title and axis labels.  
  plt.show() displays the plot.  

* BAR PLOT  
  Bar plot are used to represent data in categories. it is particularly useful when comparing quantities across different categories.

```python
categories = ['A','B','C','D','E']
values = [10,20,30,40,50]

plt.bar(categories,values,color='green', alpha=0.7)
plt.title('Bar plot example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show
```

This bar plot compares the quantities of four categories. Category D has highest value, while category A has lowest.

* HISTOGRAM  
  Histogram are used to show the distribution of data. It groups the data into bins and shows how many data points fall into each bins.

```python
import numpy as np

data = np.random.randn(1000) # Generate 1000 random values from a normal distribution.

plt.hist(data, bins=20,color='purple',edgecolor = 'black')
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

This histogram shows the distribution of generated data. Since the data is from a noemal distribution, we expect the histogram to be bell-shaped.

* SCATTER PLOT  
  Scatter plot display individual data points and are useful for understanding the relationship between two variables.

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]

plt.scatter(x,y, color='red')
plt.title('Scatter plot example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

This scatter plot shows the relationship between x and y, where y seems to increase as x increase. The points do not follow a perfectly linear relationship, indicating the data might be non-linear.

# Matplotlib Customization

Matplotlib provides several ways to customize the appearance of your plots to improve visual appear or clarity.

* Cusomizing the line style  
  You can customize the line style (solid,dashed,etc.), marker style and color.

```python
plt.plot(x,y, color='green',linestyle='--',marker='s',markersize=8)
plt.title('Custom line style')
plt.show()
```

* changing plot size  
  Use fgisize to change the sizze of your figure ( in inches).

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
plt.show()
```

* SUBPLOTS  
  You can create multiple plots within one figure subplots.

```python
# Create 1 row, 2 column of subplots
fig, axs = plt.subplots(1,2, figsize=(5,3))

# First plot line (Line plot)
axs[0].plot(x,y, 'b-')
axs[0].set_title('line plot')

# Second plot (Bar plot)
axs[1].bar(categories, values, color='orange')
axs[1].set_title("bar plot")

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

* Advanced Plot Types

* Heatmaps: 
A heatmap is a way to represent data in matrix form, where individual values are represented by colors.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data = np.random.rand(10, 10)

# Create heatmap
plt.imshow(data, cmap='coolwarm', interpolation='nearest')
plt.colorbar()  # Show color scale
plt.title('Heatmap Example')
plt.show()
```

* 3D Surface Plot : 
Matplotlib supports 3D plotting through Axes3D in the mpl_toolkits.mplot3d module.

```python
from mpl_toolkits.mplot3d import Axes3D

# Generate 3D data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create 3D surface plot
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('3D Surface Plot')
plt.show()
```

# Basic Seaborn Plot Types

* Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset('tips')

# Create line plot
sns.lineplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day")
plt.show()
```

* Explanation:  

sns.load_dataset('tips') loads the built-in tips dataset, which contains information about restaurant tips (e.g., total bill, tip amount, day, time, size of the party).  
sns.lineplot() is used to create a line plot. Here, we specify x="day" and y="total_bill", so it plots the total_bill (y-axis) over the days of the week (x-axis).  
The plt.title() function adds a title to the plot, in this case, "Total Bill by Day."  
plt.show() displays the plot

* Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')

# Scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time")
plt.title("Total Bill vs Tip")
plt.show()
```

Explanation:  

sns.scatterplot() is used here to create a scatter plot, which visualizes the relationship between total_bill (x-axis) and tip (y-axis).  
The hue="time" parameter colors the points based on whether the meal occurred during lunch ('Lunch') or dinner ('Dinner').  This allows you to see if there's a difference in tip size based on the time of day.  
plt.title() adds the title "Total Bill vs Tip" to the plot.  
plt.show() displays the plot.  

* Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')

# Box plot
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

Explanation:  

sns.boxplot() creates a box plot, where x="day" represents the days of the week on the x-axis, and y="total_bill" represents the total bill on the y-axis.  
palette="Set2" sets the color palette for the plot.  
The box plot visually shows the distribution of total_bill values for each day, with the box representing the interquartile range (IQR), the line in the middle representing the median, and the whiskers showing the range.

* Histogram and KDE Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')

# Plot histogram and KDE
sns.histplot(tips['total_bill'], kde=True, color='blue')
plt.title("Histogram with KDE")
plt.show()
```

Explanation:  

sns.histplot() is used to create a histogram of the total_bill column.  
The kde=True parameter adds a Kernel Density Estimation (KDE) curve on top of the histogram, which provides a smooth estimate of the data’s distribution.  
color='blue' sets the color of the histogram and KDE curve.  
plt.title() adds the title "Histogram with KDE."  
plt.show() displays the plot.
